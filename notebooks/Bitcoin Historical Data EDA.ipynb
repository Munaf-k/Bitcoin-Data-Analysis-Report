```python
plt.figure(figsize=(12, 6))
plt.plot(daily_data.index, daily_data['Close'], label='Daily Close Price', color='blue')
plt.plot(daily_data.index, daily_data['7-day MA'], label='7-day Moving Average', linestyle='--', color='orange')
plt.plot(daily_data.index, daily_data['30-day MA'], label='30-day Moving Average', linestyle='--', color='red')
plt.xlabel('Date')
plt.ylabel('Price (BTC)')
plt.title('Bitcoin Prices with Moving Averages')
plt.legend()
plt.show()
```
